```python
import sentencepiece as spm
import pickle
import pathlib
```

```python
path = pathlib.Path('/home/gaurav/PycharmProjects/nlp-for-gujarati/language-model')
```

```python
p = path.glob('GujaratiWikipediaArticles/*')
files = [x for x in p if x.is_file()]
```
